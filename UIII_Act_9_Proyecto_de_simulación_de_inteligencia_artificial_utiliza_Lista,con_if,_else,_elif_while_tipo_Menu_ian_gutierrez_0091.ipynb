```python
# ===================================
# FARMACEUTICO
# ===================================

farmaceuticos = []

print("===================================")
print(" FARMACEUTICO ")
print("===================================")

while True:

    print("\n=========== MENU ===========")
    print("1. Agregar medicamento")
    print("2. Ver medicamentos")
    print("3. Analizar medicamento")
    print("4. Eliminar medicamento")
    print("5. Salir")

    opcion = int(input("Seleccione una opcion: "))

    # -----------------------------------
    # OPCION 1
    # -----------------------------------

    if opcion == 1:

        medicamento = input("Ingrese el nombre del medicamento: ")

        farmaceuticos.append(medicamento)

        print("Medicamento agregado correctamente.")

    # -----------------------------------
    # OPCION 2
    # -----------------------------------

    elif opcion == 2:

        print("\nLISTA DE MEDICAMENTOS")
        print(farmaceuticos)

        if farmaceuticos == []:
            print("La lista esta vacia.")

        else:
            print("Hay medicamentos registrados.")

    # -----------------------------------
    # OPCION 3
    # -----------------------------------

    elif opcion == 3:

        buscar = input("Ingrese el medicamento a analizar: ")

        if buscar == "paracetamol":
            print("Medicamento para fiebre y dolor.")

        elif buscar == "ibuprofeno":
            print("Medicamento antiinflamatorio.")

        elif buscar == "jarabe":
            print("Medicamento para la tos.")

        else:
            print("Medicamento no registrado en el sistema.")

    # -----------------------------------
    # OPCION 4
    # -----------------------------------

    elif opcion == 4:

        eliminar = input("Ingrese el medicamento a eliminar: ")

        if eliminar in farmaceuticos:

            farmaceuticos.remove(eliminar)

            print("Medicamento eliminado correctamente.")

        else:

            print("El medicamento no existe.")

    # -----------------------------------
    # OPCION 5
    # -----------------------------------

    elif opcion == 5:

        print("\n===================================")
        print(" REPORTE FINAL ")
        print("===================================")

        print("Medicamentos registrados:")
        print(farmaceuticos)

        print("Programa realizado por Ian Gutierrez NL=26")

        print("\nGracias por utilizar FARMACEUTICO")

        break

    # -----------------------------------
    # OPCION INVALIDA
    # -----------------------------------

    else:

        print("Opcion no valida")
```
